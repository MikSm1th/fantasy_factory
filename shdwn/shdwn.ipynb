```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("contest-standings-113707720.csv", low_memory=False)
df
```

```python
def create_lineup_frame(df):
    df = df.Lineup.str.split(r'(FLEX|CPT)', expand=True)
    replace_values = {"": np.nan, " ": np.nan, "FLEX": np.nan, "CPT": np.nan}
    df.replace(replace_values, inplace=True)
    df = df.dropna(axis=1, how='all')
    df = df.dropna(axis=0, how='all')
    cols = ['FLEX', 'FLEX', 'FLEX', 'FLEX', 'FLEX', 'CPT']
    df.columns = cols
    return df

create_lineup_frame(df)
```

```python
def track_players(df):
    df = df.dropna(axis=0, subset=["Lineup"])
    df = df.EntryName.str.split("(", expand=True)
    cols = ['Player', 'Drop']
    df.columns = cols
    df = df.Player
    return df

track_players(df)
```
