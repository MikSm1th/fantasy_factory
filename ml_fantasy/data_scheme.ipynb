## Normalizing Names

```python
names_df = pd.read_csv('data/all_nfl_team_abbr.csv')
names_df.head(3)
```

```python
names_df.replace({"Oakland Raiders": "Las Vegas Raiders", \
                  "San Diego Chargers": "Los Angeles Chargers", \
                 "St.Louis / Los Angeles Rams": "Los Angeles Rams"}, inplace=True)

names_df.sort_values('TeamName', inplace=True)
names_df.drop(names_df.columns[1:], axis=1, inplace=True)
```

```python
def get_unique_team_names(df, col):
    temp_df = pd.read_csv(df) 
    temp_list = list(temp_df[col].unique())
    temp_list.sort()
    return temp_list
```

```python
import glob

temp_lists=[]
temp_cols=[]
for file in glob.glob('data/*_Fanduel_Results.csv'):
    temp_list = get_unique_team_names(file, 'Team')
    temp_col = file
    temp_cols.append(temp_col)
    temp_lists.append(temp_list)
```
