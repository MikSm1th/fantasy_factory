```python
from nameparser import HumanName
import pandas as pd
```

```python
path = "data/"

df_abbr = pd.read_csv(path + 'all_nfl_team_abbr_2021.csv')
fanduel_2020 = df_abbr['2020_Fanduel']
fanduel_2021 = df_abbr['2021_Fanduel']
statistics_2020 = df_abbr['2020_stats']
statistics_2021 = df_abbr['2021_stats']
```

```python
def normalize_team_name(df1, abbr):
    """
    Team name: from full name to abbrevaition
    """
    ref = statistics_2020
    subs = dict(zip(abbr, ref))
    df1['recent_team'] = [subs.get(item) for item in df1['recent_team']]
    try:
        df1['Oppt'] = [subs.get(item) for item in df1['Oppt']]
    except:
        pass
```
