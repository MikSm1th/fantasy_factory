```python
from nameparser import HumanName
import pandas as pd
```

```python
path = "data/"

df_abbr = pd.read_csv(path + 'all_nfl_team_abbr_2021.csv')
fanduel_2020 = df_abbr['2020_Fanduel']
fanduel_2021 = df_abbr['2021_Fanduel']
statistics_2020 = df_abbr['2020_stats']
statistics_2021 = df_abbr['2021_stats']
```

```python
years = ["2020", "2021"]
for year in years:
    df1 = pd.read_csv(path+year+"_Fanduel_Results.csv")
    df1 = df1.rename(columns = {"Week":"weeks", "Name":"name", "Year":"Season"})
    
    df2 = pd.read_csv(path+"wkly_plyr_stats_" + year + ".csv")
    df2.iloc[df2["recent_team"]=="JAX", "recent_team"] = "JAC"
    
    if(year == '2020'):
        normalize_team_name(df2, statistics_2020)
    else:
        normalize_team_name(df2, statistics_2021)
    #special_names = normalize_statistics_player_name(df2)
    
```

```python
def normalize_team_name(df1, abbr):
    """
    Team name: from full name to abbrevaition
    """
    ref = statistics_2020
    subs = dict(zip(abbr, ref))
    df1['Team'] = [subs.get(item) for item in df1['Team']]
    try:
        df1['Oppt'] = [subs.get(item) for item in df1['Oppt']]
    except:
        pass
```
