```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
from io import StringIO
```

```python
def get_weekly_fanduel_stats(week, year=2021):
    yr = year
    wk = week
    if year == 2021:
        url = f"http://rotoguru1.com/cgi-bin/fyday.pl?week={wk}&game=fd&scsv=1"
    else:
        url = f"http://rotoguru1.com/cgi-bin/fyday.pl?week={wk}&year={yr}&game=fd&scsv=1"
    html = requests.get(url).text
    page = soup(html)
    data = page.find("pre").text
    STRINGDATA = StringIO(data)
    df = pd.read_csv(STRINGDATA, sep=";")
    df.to_csv(f"data/{yr}_Fanduel_Results_{wk}", index=False)
```

```python
for i in range(1,5):
    get_weekly_fanduel_stats(i)
for i in range(1,18):
    get_weekly_fanduel_stats(i, 2020)
```

```python
for i in range(1,18):
    get_weekly_fanduel_stats(i, 2019)    
for i in range(1,18):
    get_weekly_fanduel_stats(i, 2018)
```
