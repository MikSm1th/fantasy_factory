```python
import pandas as pd
import numpy as np
```

## Normalizing Names

```python
names_df = pd.read_csv('data/all_nfl_team_abbr.csv')
names_df.head(3)
```

```python
names_df.replace({"Oakland Raiders": "Las Vegas Raiders", \
                  "San Diego Chargers": "Los Angeles Chargers", \
                 "St.Louis / Los Angeles Rams": "Los Angeles Rams"}, inplace=True)

names_df.sort_values('TeamName', inplace=True)
names_df.drop(names_df.columns[1:], axis=1, inplace=True)
names_df1 = names_df.reset_index(drop=True)
#names_df1
```

```python
def get_unique_team_names(df, col):
    temp_df = pd.read_csv(df) 
    temp_list = list(temp_df[col].unique())
    temp_list.sort()
    return temp_list
```

```python
import glob

temp_lists=[]
temp_cols=[]
for file in glob.glob('data/202*_Fanduel_Results.csv'):
    temp_list = get_unique_team_names(file, 'Team')
    temp_col = file
    temp_cols.append(temp_col)
    temp_lists.append(temp_list)
```

```python
print(temp_lists, temp_cols)
```

```python
names_df2 = pd.DataFrame(temp_lists).T
temp_cols_short = [i[5:17] for i in temp_cols]
names_df2.columns = temp_cols_short
#names_df2
```

```python
wk_temp = pd.read_csv('data/wkly_plyr_stats_2020.csv')
#wk_temp.columns
```

```python
team_list = list(wk_temp.recent_team.unique())
team_list.sort()
print(team_list)
```

```python
names_df = pd.concat([names_df1, names_df2], axis=1)
names_df['2020_stats'] = team_list
names_df.head(3)
```

```python
wk_temp = pd.read_csv('data/wkly_plyr_stats_2021.csv')
#wk_temp.columns
```

```python
team_list = list(wk_temp.recent_team.unique())
team_list.sort()
print(team_list)
```

```python
names_df['2021_stats'] = team_list
names_df
```
