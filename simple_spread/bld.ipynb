```python
#%%writefile spread_scrape.py
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
def get_weekly_spread():
    url = "https://rotogrinders.com/lineups/nfl"
    html = requests.get(url).text
    page = soup(html)
    return page
```

```python
page = get_weekly_spread()
print(page)
```
