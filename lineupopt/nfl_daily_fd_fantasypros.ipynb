```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
df_fd_temp = pd.read_csv("data/nfl/FanDuel-NFL-2021 ET-10 ET-17 ET-65217-players-list.csv")
df_fd = df_fd_temp.convert_dtypes()
df_fd.head()
```

```python
import glob

df_list = []
for file in glob.glob('data/nfl/FantasyPros_Fantasy_Football_Projections_*.csv'):
    temp_df = pd.read_csv(file)
    #print(temp_df.columns)
    temp_df = temp_df[['Player', 'FPTS']]
    df_list.append(temp_df)
    
proj_df = pd.concat(df_list, axis=0)
proj_df.dropna(inplace=True)

proj_df
```

```python
replace_dict = {"Patrick Mahomes II":"Patrick Mahomes", }
clean_players = [replace_dict.get(item,item) for item in proj_df.Player]
proj_df['Player'] = clean_players
```

```python
df_fd = df_fd.merge(proj_df, how='left', left_on='Nickname', right_on='Player')
df_fd
```

```python
df_fd["FPPG"] = df_fd["FPTS"]
df_fd.drop(labels = ["Player", "FPTS"], axis=1, inplace=True)
df_fd.head(3)
```

```python
df_fd.to_csv('data/nfl/FDprojections.csv')
```

```python
from pydfs_lineup_optimizer import Site, Sport, get_optimizer, PlayerFilter, AfterEachExposureStrategy, TeamStack

optimizer = get_optimizer(Site.FANDUEL, Sport.FOOTBALL)
optimizer.load_players_from_csv("data/nfl/FDprojections.csv")
optimizer.add_stack(TeamStack(3, for_positions=['QB', 'WR', 'TE']))
#optimizer.add_stack(TeamStack(2, for_positions=['RB', 'D']))
for lineup in optimizer.optimize(40, max_exposure=0.4, exposure_strategy=AfterEachExposureStrategy):
    print(lineup)
optimizer.export('data/nfl/fd_result.csv')
```
